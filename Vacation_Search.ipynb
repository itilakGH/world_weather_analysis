```python
# Importing the dependencies.
import pandas as pd
import gmaps
import requests
# Import the API key.
from config import g_key
gmaps.configure(api_key=g_key)
```

```python
# Importing the CSV created in part one into a DataFrame.
city_data_df = pd.read_csv("weather_data/WeatherPy_Challenge.csv")
city_data_df.head()
```

```python
# Adding columns to map raining and snowing conditions
def raining(col):
    if col > 0:
        return 'yes'
    else:
        return 'no'
    
def snowing(col):
    if col > 0:
        return 'yes'
    else:
        return 'no'
```

```python
# Mapping columns
city_data_df['raining']=city_data_df['Rain inches last 3 hours'].apply(raining)
city_data_df['snowing']=city_data_df['Snow inches last 3 hours'].apply(raining)
```

```python
city_data_df.head(10)
```

```python
# Creating input fields to ask the customer to add preferences.
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))
rain_ask = str(input("Do you want it to be raining? (yes/no) "))
snow_ask = str(input("Do you want it to be snowing? (yes/no) "))
```

```python
# Filter the dataset to find the cities that fit the criteria.
preferred_cities_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp) & \
                                       (city_data_df["Max Temp"] >= min_temp) &
                                       (city_data_df["raining"] == rain_ask) &
                                       (city_data_df["snowing"] == snow_ask)]
preferred_cities_df.head(10)
```

```python
# Creating a new DataFrame with hotels to store hotel names along with existing data.
hotel_df = preferred_cities_df[["City", "Country", "Max Temp", "Lat", "Lng", "Current Description"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# Setting parameters to search for a hotel.
params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key
}
```

```python
# Iterate through the DataFrame.
for index, row in hotel_df.iterrows():
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to location key for the params dictionary.
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # Make request and get the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
       
    # Grab the first hotel from the results and store the name.
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except (IndexError):
        print("Hotel not found... skipping.")
```

```python
hotel_df.head()
```

```python
# Defining the info in the pop-up marker
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
<dt>Current Weather</dt><dd>{Current Description}</dd>
</dl>
"""
```

```python
# Storing the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
hotel_info
```

```python
# Adding a heatmap of temperature for the vacation spots and a pop-up marker for each city.
locations = hotel_df[["Lat", "Lng"]]
max_temp = hotel_df["Max Temp"]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
heat_layer = gmaps.heatmap_layer(locations, weights=max_temp,dissipating=False,
             max_intensity=300, point_radius=4)
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```

```python
# Creating the output file (CSV).
output_data_file = "weather_data/WeatherPy_Vacation.csv"
# Export the City_Data into a CSV.
hotel_df.to_csv(output_data_file, index_label="City_ID")
```
