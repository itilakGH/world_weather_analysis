```python
# Importing the dependencies.
import pandas as pd
from datetime import datetime
import numpy as np
import requests
from config import weather_api_key
from citipy import citipy
```

```python
# Creating a set of random latitude and longitude combinations.
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)
lat_lngs 
```

```python
# Adding the latitudes and longitudes to a list.
coordinates = list(lat_lngs)
```

```python
# Starting URL for Weather Map API Call.
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
print(url)
```

```python
# Creating a list for holding the cities.
cities = []
# Identifying the nearest city for each latitude and longitude combination.
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
# Printing the city count to confirm sufficient count.
len(cities)
```

```python
# Creating an empty list to hold the weather data.
city_data = []
# Printing the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Creating counters.
record_count = 1
set_count = 1
```

```python
# Looping through all the cities in the list.
for i, city in enumerate(cities):

    # Groupping cities in sets of 50 for logging purposes.
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
    # Creating endpoint URL with each city.
    city_url = url + "&q=" + city

    # Logging the URL, record, and set numbers and the city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    # Adding 1 to the record count.
    record_count += 1
    # Running an API request for each of the cities.
    try:
        # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()
        # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_desc = city_weather['weather'][0]['description']
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        try:
            # Because the key of the dictionary can be 1h or 3h we cannot be specific 
            # so we have to take only the value of the first key in the list of dictionaries
            city_rain = city_weather["rain"][list(city_weather['rain'].keys())[0]]
        except:
            city_rain = 0
        
        try:
            # Because the key of the dictionary can be 1h or 3h we cannot be specific 
            # so we have to take only the value of the first key in the list of dictionaries
            city_snow = city_weather["snow"][list(city_weather['snow'].keys())[0]]
        except:
            city_snow = 0
            
        # Convert the date to ISO standard.
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        # Append the city information into city_data list.
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Current Description": city_desc,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date,
                          "Rain inches last 3 hours": city_rain,
                          "Snow inches last 3 hours": city_snow})

# If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
        pass

# Indicating that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
# Converting the array of dictionaries to a DataFrame.
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)
```

```python
# New column order
new_column_order = ['City','Country','Date','Lat','Lng','Max Temp','Humidity','Cloudiness','Wind Speed',
                    'Current Description','Rain inches last 3 hours','Snow inches last 3 hours']
city_data_df = city_data_df[new_column_order]
city_data_df.head(10)
```

```python
# Filtering and counting the number of cities with recorded rain or snow
count_data_df = pd.DataFrame(city_data_df[(city_data_df["Rain inches last 3 hours"] > 0) | (city_data_df["Snow inches last 3 hours"] > 0)])
count_data_df.count()["City"]
```

```python
# Creating the output file (CSV).
output_data_file = "weather_data/WeatherPy_Challenge.csv"
# Export the City_Data into a CSV.
city_data_df.to_csv(output_data_file, index_label="City_ID")
```
